# Combined geospatial files so we can assign a nyct to each ride

```python
# import needed files
import geopandas as gpd
nyct2010_df=gpd.read_file('Data_Sources\\nyct2010_25c')
nyct2020_df=gpd.read_file('Data_Sources\\nyct2020_25c')
modzcta_df=gpd.read_file('Data_Sources\\Modified Zip Code Tabulation Areas (MODZCTA)_20251024')
nta_df=gpd.read_file('Data_Sources\\2020 Neighborhood Tabulation Areas (NTAs)_20251024')
```

```python
# using nyct2020_zcta as the base and boroct2020 as an index, merge by nta2020
merged_df=pd.merge(nyct2020_df,nta_df,how='left',left_on='NTA2020',right_on='nta2020',suffixes=['_nyct','_nta'])
merged_df.info()
```

```python
# now filter out only manhattan
manhattan_nyct_nta=merged_df[merged_df['BoroName']=='Manhattan']
manhattan_nyct_nta.info()
manhattan_nyct_nta.head()
```

```python
# using boroct2020 as an index
# we want to keep BoroCT2020, NTAName, NTA2020, GEOID, geometry_nyct,geometry_nta
manhattan_nyct_nta_filtered = manhattan_nyct_nta[
    ['BoroCT2020', 'NTAName', 'NTA2020', 'GEOID', 'geometry_nyct', 'geometry_nta']
].set_index('BoroCT2020')
manhattan_nyct_nta_filtered.info()


# now we should be able to join the modzta onto it (I HOPE???)
manhattan_nyct_nta_filtered.head()
```

```python
# create new file with exploded zcta with corresponding modzcta
modzcta_zcta=modzcta_df.copy()

# make each zcta it's own column
col_name='zcta'
modzcta_zcta=modzcta_zcta.assign(**{col_name:modzcta_zcta[col_name].str.split(',')})
modzcta_zcta=modzcta_zcta.explode('zcta',ignore_index=True)
display(modzcta_zcta.head())
modzcta_zcta.info()
```

```python
# drop central park
modzcta_zct_fil = modzcta_zcta[modzcta_zcta['modzcta']!='99999'] #remove 99999 central park
modzcta_zct_fil.info()
modzcta_zct_fil[modzcta_zct_fil['modzcta']=='99999']
```

```python
#trip_copy_df['geometry']=gpd.points_from_xy(trip_copy_df['pickup_longitude'], trip_copy_df['pickup_latitude'],crs="EPSG:4326")
#df_gpd=gpd.GeoDataFrame(data=trip_copy_df, geometry='geometry')

manhattan_gdf=gpd.GeoDataFrame(data=manhattan_nyct_nta_filtered,geometry='geometry_nyct',crs="EPSG:2263")
#manhattan_gdf.info()
manhattan_gdf.to_crs("EPSG:4326",inplace=True)

manhattan_gdf.head()
#spatially merge modzcta onto it now
assigned_gdf=gpd.sjoin(manhattan_gdf,modzcta_zct_fil,how='left')
assigned_gdf.info()
assigned_gdf.head()
assigned_gdf['BoroCT2020_col']=assigned_gdf.index
assigned_gdf.head()
assigned_unique_gdf=assigned_gdf.drop_duplicates(subset='BoroCT2020_col',keep='last')
assigned_unique_gdf.info()
assigned_unique_gdf.head()
# we only want to keep geometry_nyct, drop geometry_nta
assigned_unique_gdf.drop(columns=['geometry_nta','index_right'],inplace=True)
assigned_unique_gdf.head()
```

```python
# manually assign NTA2020 code of MN0191 to the modzcta of 10004
ellis_island=modzcta_zcta[modzcta_zcta['modzcta']=='10004']


df=assigned_unique_gdf.copy()
cols = ["modzcta", "label", "zcta", "pop_est"]
df.loc[df.index[0], cols] = ellis_island.loc[ellis_island.index[0], cols].values
```
